```python
import sympy

# f(x) =0.6x^3+5.5x^2+10x -5

x, a, b, c, d = sympy.symbols('x, a, b, c, d')
params = {a: 0.6, b: 5.5, c: 10, d: -5}

f = a * (x ** 3) + b * (x ** 2) + c * x + d
f_pr = f.subs(params)

print(f'f(x) = {f_pr}')
roots = sympy.solve(f_pr, x)
for i in range(len(roots)):
    print(f'Корень {i + 1}: {roots[i].n(3)}')
```

```python
gr1 = sympy.plotting.plot(f_pr, (x, -10, 5), title='График функции')
```

```python
df = sympy.diff(f, x)
df_pr = df.subs(params)
print(f"f'(x) = {df_pr}")
gr2 = sympy.plotting.plot(df_pr, (x, -10, 5), title='График производной функции')
```

```python
gr3 = sympy.plotting.plot(f_pr, df_pr, (x, -10, 5), title='Совмещенный график', legend=True, xlabel="X", ylabel="Y")
```

```python
df_roots = sympy.solve(df_pr, x)
for i in range(len(df_roots)):
    print(f'Вершина {i + 1} графика функции: {df_roots[i].n(3)}')
```

```python
print('Функция возрастает на промежутках:')
sympy.solve(df_pr > 0, x)
```

```python
print('Функция убывает на промежутках:')
sympy.solve(df_pr < 0, x)
```

```python
print('Функция больше нуля на промежутках:')
sympy.solve(f_pr > 0, x)
```

```python
print('Функция меньше нуля на промежутках:')
sympy.solve(f_pr < 0, x)
```
